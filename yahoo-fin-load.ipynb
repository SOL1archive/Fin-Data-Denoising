```python
import yfinance as yf

snp500_df =  yf.download(
    tickers="^GSPC", 
    start="1990-02-24", 
    end="2023-02-24", 
    interval="1d"
    )
# including Dot-com bubble (1995-2002), Great Recession(2008-2011), Great Lockdown(2020-2023)

snp500_df.head()
```

```python
snp500_df['mean'] = snp500_df[['Open', 'High', 'Low', 'Close']].mean(axis=1)
snp500_df.plot.line(y='mean')
```

```python
snp500_df.to_csv('snp.csv')
```
