```python
import yfinance as yf

import pandas as pd
from yahoo_fin_api import load_df
```

## Download and Set Mean

```python
df =  yf.download(
    tickers="^TNX", 
    start="1990-02-24", 
    end="2023-02-24", 
    interval="1d"
    )
# including Dot-com Bubble (1995-2001), Great Recession(2008-2011), Great Lockdown(2020-2023)

df.head()
```

```python
df['Mean'] = df[['Open', 'High', 'Low', 'Close']].mean(axis=1)
df.head()
```

```python
FILE_NAME = 'treasury-yield-10y'
```

```python
df.to_csv(f'data/{FILE_NAME}.csv')
```

```python
df = load_df(f'data/{FILE_NAME}.csv')
df.head()
```

## Minor Viz

```python
fig = df.plot.line(y='Mean').get_figure()
fig.set_size_inches(18, 10)
```
